```python
df.drop(["Date"],axis=1,inplace=True)
df.shape
y = df["Adj Close"].values
```

```python
df.drop(["Adj Close"],axis=1,inplace=True)
```

```python
x = df.iloc[:,:]
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
x = ss.fit_transform(x)
y = y.reshape(-1,1)
y = ss.fit_transform(y)
```

```python
import sklearn
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
 x_train
```

```python
y_train
```

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
linear = list(cross_val_score(LinearRegression(),x,y,cv=10))
print("Cross Validation Score of linear : "+str(np.average(linear)) )

rf = list(cross_val_score(RandomForestRegressor(),x,y,cv=10))
print("Cross Validation Score of rf : "+str(np.average(rf)))

dt = list(cross_val_score(DecisionTreeRegressor(),x,y,cv=10))
print("Cross Validation Score of dt : "+str(np.average(dt)))
```

```python
y.dtype
```

```python
x.dtype
```

```python
model_para = {
    "linear":{
        "model":LinearRegression(),
        "para":{}
    },
    "rf":{
        "model":RandomForestRegressor(),
        "para":{
            'n_estimators': [1,5,6,7,8,9,10,20,30,50,100,200],
             'criterion':["squared_error", "absolute_error", "poisson"]
        }
    },
    "dt":{
        "model":DecisionTreeRegressor(),
        "para":{
            "criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"]
        }
    }
}
```

```python
scores = []
for model_name,mp in model_para.items():
    print(model_name)
    print(mp)
    clf = GridSearchCV(mp["model"],mp["para"],cv=10)
    clf.fit(x,y)
    scores.append({
        "model_name":model_name,
        "best score":clf.best_score_,
        "best parameters":clf.best_params_
    }
        
    )

#print(scores)
```

```python
scores
```
